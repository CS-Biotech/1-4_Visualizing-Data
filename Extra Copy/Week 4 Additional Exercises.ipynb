In the code below, we are once again generating the heatmap from the main module.

````python
import pandas as pd
import seaborn as sns

df = pd.read_csv('hf_data_tut.csv')


#TODO: Add additional customization below

# generate the heatmap
heatmap = sns.heatmap(df.corr(), annot=True, cmap='Reds')
# Resize the plot for better viewing
heatmap.figure.set_figwidth(15)
heatmap.figure.set_figheight(10)

````

This plot effectively serves our analytical needs, as we are familiar with the context of the data and the specific questions it helps us address. Yet, for others lacking this background, the significance of the heatmap might not be immediately clear. To bridge this gap, we'll enhance the diagram with additional informative elements, ensuring it conveys the necessary context and insights more clearly to all viewers. **Starting off, in the code cell above, lets add a title.** You can do so with the ```set_title()``` function.

Now without altering the title and keeping the default parameters, our title may look a bit small and insignificant. It would be nice to see if we can change the characteristics of the title and adjust its font size to be bigger. To do so, you will have to change the argument ```fontdict``` in the function. As the name suggests, this parameter takes a dictionary of characteristics to the font that will be used for the title. We are specifically interested in ```fontsize```. Therefore, to accomplish this, **please set the ```'fontsize'``` key in the ```fontdict``` parameter to the value 20**. 

As a final adjustment to the title of the heatmap, we would like to move the title a bit farther away from the heatmap itself. To do this you will have to **adjust the ```pad``` parameter in ```set_title()```; set it to 15**.
