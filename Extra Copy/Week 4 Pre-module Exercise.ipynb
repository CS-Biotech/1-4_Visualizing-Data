## Histograms

We learned some valuable information from descriptive statistics in the previous week, but it is difficult to get a sense of the distribution of data just by looking at the percentiles generated by by the DataFrame function ```describe()```. We can leverage visualization functions to generate histograms of the data so we can better understand the distribution in one snapshot.

Pandas DataFrames have a convenient function for this purpose, called ```hist()```. This function will generate a histogram for each column in your DataFrame with the default style settings. It has many optional parameters that allow you to specify the number of buckets, the columns you want (if not all), the size and labels of the plots, and more. 


| Function | Input parameters | Output | Syntax |
| --- | --- | --- | --- |
| hist() | n/a (see documentation for optional parameters) | A histogram for each column in the DataFrame. | df.hist() |

````python
df.hist(figsize=(8,10))
````
